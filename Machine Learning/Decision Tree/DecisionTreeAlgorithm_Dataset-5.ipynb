# converting Target Object into integer 

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_test = encoder.fit_transform(target)
# Print the original and encoded target values
print("Original target values:")
print(target)
print("\nEncoded target values:")
print(y_test)
```

```python
# Separating the target variable
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

```

```python
X = data.iloc[:, : -1].values     #all the columns except the last column, dropping the last columm
y = np.array(data.iloc[:, -1].values)  
```

```python
#dividing the data inot test and train and assigned  20% of the data for testing
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size= 0.20, random_state= 1)
X_train, x_test, y_train, y_test
```

```python
#importing standard scalling method from sklearn
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
#providing the inputs for the scalling purpose
X_train = sc.fit_transform(X_train)
x_test = sc.transform(x_test)
X_train, x_test
```

```python

# importing decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
from inequality import gini
# Creating the classifier object
clf= DecisionTreeClassifier(criterion = "gini", random_state = 100) #ValueError: Unknown label type: 'continuous'
# providing the training dataset
clf.fit(X_train,y_train)
```

```python
y_pred= clf.predict(x_test)
```

```python
# importing the plot tree method
from sklearn.tree import DecisionTreeClassifier, plot_tree
clf = DecisionTreeClassifier()
# output size of decision tree
plt.figure(figsize=(40,20))
# providing the training dataset
clf = clf.fit(X_train, y_train)
plot_tree(clf, filled=True)
plt.title("Decision tree training for training dataset")
plt.show()
```

```python
# drawing the tree in the form of text to make it easier to understand and read
from sklearn import tree
text_representation = tree.export_text(clf)
print(text_representation)
```
